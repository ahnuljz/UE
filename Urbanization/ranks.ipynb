```python
import pandas as pd
import numpy as np

euc = pd.read_excel('D:/ee-composite/src/sqlserver/data/CCD/Data_ccd_city_mean.xlsx').set_index('sn')
slope = pd.read_excel('D:/ee-composite/src/sqlserver/data/CCD/Data_ccd_slope.xlsx').set_index('sn')
# print(euc,euc.shape)

euc['slope_e'] = slope['slope_e']
euc['slope_u'] = slope['slope_u']
euc['slope_ccd'] = slope['slope_ccd']
# print(euc,euc.shape)

#排名
ccdrank = euc.loc[:,['city','E','U','ccd','slope_ccd']]
ccdrank['rankccd'] = ccdrank['ccd'].rank(method='first',ascending=False)
ccdrank['ranke'] = ccdrank['E'].rank(method='first',ascending=False)
ccdrank['ranku'] = ccdrank['U'].rank(method='first',ascending=False)
print(ccdrank,ccdrank.shape)
ccdrank.to_excel('Data_ccd_ranks_mean.xlsx', sheet_name='rank')
```
